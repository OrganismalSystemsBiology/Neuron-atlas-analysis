```python
for args0 in args0s:
    print(args0)
    t_start = time.time()

    ######## __Load JSON__#######
    param_path = args0
    with open(param_path) as f:
        param_stitch = json.load(f)

    ####### ___List of Operations___##########
    
    if 1==1: # image stitching only
        command = {"check": False,  
                   "xy_stitching":True,
                   "FW_RV_stitching": True, #True
                   "wrong_FW_RV_stitch_stacks": param_stitch["FW_RV_stitching"]["wrong_FW_RV_stitch_stacks"],
                   "image_making":
                   {"same_size_with_density_image": True,
                    "wrong_FW_RV_stitch_stacks_checker": True, #True # False:only making whole image
                    "average_image": 0,# Enter 0 if you do not want to create it
                    "normalized_image": 0,# Enter 0 if you do not want to create it
                    "down_norishiro": 0},
                   "point_transfer": False, 
                   "image_size": 2060,
                   "FW_RV_rasen": 25,
                   "density_image_voxel_size_um": 30,
                   "FW_RV_accuracy_cut_off": 0.8}
    
    if 1==0: # point transfer only
        command = {"check": False,  
                   "xy_stitching":False, #True
                   "FW_RV_stitching": False, #True
                   "wrong_FW_RV_stitch_stacks": param_stitch["FW_RV_stitching"]["wrong_FW_RV_stitch_stacks"],
                   "image_making":
                   {"same_size_with_density_image": False,#True
                    "wrong_FW_RV_stitch_stacks_checker": False, #True # False:only making whole image
                    "average_image": 0,# Enter 0 if you do not want to create it
                    "normalized_image": 0,# Enter 0 if you do not want to create it
                    "down_norishiro": 0},
                   "point_transfer": True, 
                   "image_size": 2060,
                   "FW_RV_rasen": 25,
                   "density_image_voxel_size_um": 30,
                   "FW_RV_accuracy_cut_off": 0.8}

    ###### __Initialize the class___######
    if command["image_size"] == 2060:
        stitching_class = stitching_class_2060
    #if command["image_size"] == 2048: # 2048 version 
    #    stitching_class = stitching_class_2048

    stitching_img_um = command["density_image_voxel_size_um"]

    ST_xy = stitching_class.stitching_xy(param_stitch, stitching_img_um)
    ST_FW_RV = stitching_class.stitching_FW_RV(param_stitch, stitching_img_um)

    ST_FW_RV.rasen_check_range = command["FW_RV_rasen"]

    ST_FW_RV.FW_RV_accuracy_cut_off = command["FW_RV_accuracy_cut_off"]
    ST_FW_RV.down_norishiro = command["image_making"]["down_norishiro"]

    ################################
    ########### _Check_#############
    ################################
    if command["check"] == True:
        print("###########_Check_##############")

        def make_FW_check_20(Z):
            ST_xy.check_img_ZZ(20, Z, "FW")
            return

        def make_RV_check_20(Z):
            ST_xy.check_img_ZZ(20, Z, "RV")
            return

        # self.FW_z_max represents the maximum Z in the FW stack, and self.RV_z_max represents the maximum Z in the RV stack
        larges_ = joblib.Parallel(n_jobs=45, verbose=1)(joblib.delayed(
            make_FW_check_20)(z) for z in range(100, ST_xy.FW_z_max+100, 100))
        larges_ = joblib.Parallel(n_jobs=45, verbose=1)(joblib.delayed(
            make_RV_check_20)(z) for z in range(100, ST_xy.RV_z_max+100, 100))

    ################################
    ######## _xy_stitching_#########
    ################################
    if command["xy_stitching"] == True:
        print("########_xy_stitching_##########")
        # self.FW_z_max is the maximum Z value in the FW stack, and self.RV_z_max is the maximum Z value in the RV stack
        FW_z_step = int(
            ST_xy.FW_z_max/ST_xy.param["xy_stitching"]['stitching_times'])
        FW_z_steps = FW_z_step * \
            (np.arange(ST_xy.param["xy_stitching"]
             ['stitching_times'])) + ST_xy.FW_boundary
        RV_z_step = int(
            ST_xy.RV_z_max/ST_xy.param["xy_stitching"]['stitching_times'])
        RV_z_steps = RV_z_step * \
            (np.arange(ST_xy.param["xy_stitching"]
             ['stitching_times'])) + ST_xy.RV_boundary

        print("calcurating FW shift......")
        ut1 = time.time()
        for plane in FW_z_steps:
            ST_xy.xy_stitcing_plane_parallel(plane, "FW")
        ut2 = time.time()
        print("FW shift calcurating takes {} s".format(str(ut2-ut1)))

        print("calcurating RV shift......")
        ut1 = time.time()
        for plane in RV_z_steps:
            ST_xy.xy_stitcing_plane_parallel(plane, "RV")
        ut2 = time.time()
        print("RV shift calcurating takes {} s".format(str(ut2-ut1)))
        
        # Calculate and save shift_FW and shift_RV
        ST_FW_RV.clc_shift_FW_and_shift_RV()
        ST_FW_RV.check_stitch_FW_boundary(ST_FW_RV.FW_boundary)
        ST_FW_RV.check_stitch_RV_boundary(ST_FW_RV.RV_boundary)

    ################################
    ######## _FW_RV_stitching_######
    ################################
    if command["FW_RV_stitching"] == True:
        print("########_FW_RV_stitching_#######")
        if os.path.exists(os.path.join(ST_FW_RV.stitching_check_path, "shift_FW.npy")) * os.path.exists(os.path.join(ST_FW_RV.stitching_check_path, "shift_FW.npy")) == 0:
            ST_FW_RV.clc_shift_FW_and_shift_RV()
        # Calculate the position of each stack with the top-left corner of the top-left stack as the origin
        print("load shif_FW and shift_RV...")
        ST_FW_RV.shift_FW = np.load(os.path.join(
            ST_FW_RV.stitching_check_path, "shift_FW.npy"))
        ST_FW_RV.shift_RV = np.load(os.path.join(
            ST_FW_RV.stitching_check_path, "shift_RV.npy"))
        ST_FW_RV.get_RV_coodinate_before_rotation()

        # Calculate the rotation angle of the RV
        # save rotation center and RV_boundary_hosei(for RV points rotation)
        ST_FW_RV.calc_large_rotation_transform()

        # Calculate the number of slices required to create large_RV
        ST_FW_RV.calc_RV_maisuu()

        # RV_image_rotation_5um (Preparation for creating large_RV)
        resize_um = ST_FW_RV.FW_RV_stitching_scale
        big_z_start = ST_FW_RV.RV_boundary_hosei - ST_FW_RV.RV_maisuu_half_5um * \
            int(ST_FW_RV.FW_RV_stitching_scale/ST_FW_RV.scale_z)
        big_z_end = ST_FW_RV.RV_boundary_hosei + ST_FW_RV.RV_maisuu_half_5um * \
            int(ST_FW_RV.FW_RV_stitching_scale/ST_FW_RV.scale_z) + 1
        make_z_list = list(range(big_z_start, big_z_end, int(
            ST_FW_RV.FW_RV_stitching_scale/ST_FW_RV.scale_z)))

        print("big_z_start = ", big_z_start)

        # Create a function in switching_xy_2060.py (make_big_image_RV_norm) to generate a big RV image at the z-coordinate of the rotation center

        # RV_image_rotation_5um
        resize_um = ST_FW_RV.FW_RV_stitching_scale
        big_z_start = ST_FW_RV.RV_boundary_hosei - ST_FW_RV.RV_maisuu_half_5um * \
            int(ST_FW_RV.FW_RV_stitching_scale/ST_FW_RV.scale_z)
        big_z_end = ST_FW_RV.RV_boundary_hosei + ST_FW_RV.RV_maisuu_half_5um * \
            int(ST_FW_RV.FW_RV_stitching_scale/ST_FW_RV.scale_z) + 1

        make_z_list = list(range(big_z_start, big_z_end, int(
            ST_FW_RV.FW_RV_stitching_scale/ST_FW_RV.scale_z)))
        print("load_RV_large_images_for_making_rotated_RV_large_image...")
        bigs = joblib.Parallel(n_jobs=-1, verbose=1)(joblib.delayed(ST_FW_RV.make_big_image_RV_norm)(ZZZ, resize_um)
                                                     for ZZZ, resize_um in zip(make_z_list, [ST_FW_RV.FW_RV_stitching_scale]*len(make_z_list)))
        bigs = np.array(bigs)
        print("transform_RV_img_to_ants_format...")
        niiRV_img = tif2nii_no_save(bigs, resize_um, resize_um, resize_um)
        del bigs
        gc.collect()
        antsRV_img = ants.utils.nifti_to_ants(niiRV_img)
        del niiRV_img
        gc.collect()
        transform_large_img_path = os.path.join(
            ST_FW_RV.stitching_dst_path, "large_image.mat")
        sitk.WriteTransform(ST_FW_RV.Transform_large_Zum(
            big_z_start*ST_FW_RV.scale_z, big_z_start*ST_FW_RV.scale_z), transform_large_img_path)
        print("rotation_of_RV_large_images...")
        W_img = ants.apply_transforms(
            fixed=antsRV_img, moving=antsRV_img, transformlist=transform_large_img_path)
        W_img_np = ST_FW_RV.ants2numpy(W_img)
        del antsRV_img, W_img

        large_image_boundary = (
            W_img_np[ST_FW_RV.RV_maisuu_half_5um]).astype("float32")
        tifffile.imsave(os.path.join(ST_FW_RV.stitching_dst_path,
                        "rotated_large_image.tif"), large_image_boundary)
        large_image_boundary_065 = cv2.resize(large_image_boundary, dsize=None, fx=ST_FW_RV.FW_RV_stitching_scale /
                                              ST_FW_RV.scale_x, fy=ST_FW_RV.FW_RV_stitching_scale/ST_FW_RV.scale_y)
        del W_img_np, large_image_boundary

        # Determine the z-direction scan range for FW_RV_stitching
        ST_FW_RV.FW_RV_stitching_z_search_range = 80

        # Create the function def getRVtarget(XX, YY, large_image_boundary_065) (move to switching_xy_2060.py)
        # Create the function def stitchable_FW_RV(XXX, YYY) (move to switching_xy_2060.py)

        def FW_RV_stitching(stack_num):
            YY, XX = divmod(stack_num, ST_FW_RV.x_list_num)
            z_search_range = ST_FW_RV.FW_RV_stitching_z_search_range
            z_search_maisuu = 2*z_search_range + 1
            if not ST_FW_RV.stitchable_FW_RV(XX, YY, large_image_boundary_065):
                Result_FW_RV = np.array([np.nan, np.nan, np.nan, np.nan])
            else:
                RV_img = ST_FW_RV.getRVtarget(XX, YY, large_image_boundary_065)[
                    0].astype(np.float32)
                FW_imgs = np.array([ST_FW_RV.norm_large(XX, YY, ZZZ, "FW") for ZZZ in range(
                    ST_FW_RV.FW_boundary-z_search_range, ST_FW_RV.FW_boundary+z_search_range+1)]).astype(np.float32)
                R = []
                for i, template_img in enumerate(FW_imgs):
                    # Image Search (Template Matching)
                    result = cv2.matchTemplate(
                        RV_img, template_img, cv2.TM_CCORR_NORMED)
                    # Get the confidence score and position coordinates from the search results
                    R.append(cv2.minMaxLoc(result))
                Sz = (np.argmax(np.array(R).T[1]))
                Sx, Sy = R[Sz][3]
                FW_x_hidariue, FW_y_hidariue, FW_z = ST_FW_RV.getRVtarget(XX, YY, large_image_boundary_065)[1] + Sx, ST_FW_RV.getRVtarget(
                    XX, YY, large_image_boundary_065)[2] + Sy, Sz+ST_FW_RV.FW_boundary-ST_FW_RV.FW_RV_stitching_z_search_range
                Result_FW_RV = np.array(
                    [FW_x_hidariue, FW_y_hidariue, FW_z, np.max(np.array(R).T[1])])
                img_shape = FW_imgs[Sz].shape
                save_img_path = os.path.join(
                    ST_FW_RV.stitching_check_path, "FW_RVtarget/X{0}_Y{1}.tif".format(str(XX).zfill(4), str(YY).zfill(4)))
                print(save_img_path)
                tifffile.imsave(
                    save_img_path, (RV_img[Sy:Sy+img_shape[0], Sx:Sx+img_shape[1]]*1000).astype(np.uint16))
                save_img_path = os.path.join(
                    ST_FW_RV.stitching_check_path, "FW_RVresult/X{0}_Y{1}.tif".format(str(XX).zfill(4), str(YY).zfill(4)))
                tifffile.imsave(
                    save_img_path, (FW_imgs[Sz]*1000).astype(np.uint16))
            save_path = os.path.join(os.path.join(
                ST_FW_RV.stitching_dst_path, "FW_RV_stitching_result/"), "X"+str(XX).zfill(5)+"_Y"+str(YY).zfill(5))
            np.save(save_path, Result_FW_RV)
            return

        print("FW_RV_stitching...")
        UT1 = time.time()
        FW_RV_stitching_ = joblib.Parallel(n_jobs=20, verbose=1)(joblib.delayed(FW_RV_stitching)(
            StackNum) for StackNum in range(ST_FW_RV.x_list_num*ST_FW_RV.y_list_num))
        print("FW_RV_stitching takes ", (time.time()-UT1), " s")

    ##########################################
    # _making img(same_size_with_density_img)_#
    ##########################################
    # Fill in the areas that could not be calculated during FW_RV_stitching
    if command["image_making"]["same_size_with_density_image"] == True or command["image_making"]["wrong_FW_RV_stitch_stacks_checker"] == True:
        # Read shift_FW
        print("########_making_img_#######")
        print("load shif_FW and shift_RV...")
        ST_FW_RV.shift_FW = np.load(os.path.join(
            ST_FW_RV.stitching_check_path, "shift_FW.npy"))
        ST_FW_RV.shift_RV = np.load(os.path.join(
            ST_FW_RV.stitching_check_path, "shift_RV.npy"))

        # Start creating the FW image from here
        Resize_um = command["density_image_voxel_size_um"]

        if command["image_making"]["wrong_FW_RV_stitch_stacks_checker"] == True:
            print(
                "#_making img(same_size_with_density_img)_###_wrong_FW_RV_stitch_checker##")

            print("Fill in the areas that could not be calculated during FW_RV_stitching")
            # ST_FW_RV.fill_not_calculated_FW_RV_stitching_stack(ST_FW_RV.shift_FW)
            FW_RV_nozoku_list = command["wrong_FW_RV_stitch_stacks"]
            print(FW_RV_nozoku_list)
            ST_FW_RV.fill_not_calculated_FW_RV_stitching_stack_nozoku(
                ST_FW_RV.shift_FW, FW_RV_nozoku_list)
            # Read FW_RV stitching results

            def FW_RV_stitching_data(XX, YY):
                return ST_FW_RV.FW_RV_stitching_data_np[XX, YY]

            shift_FW_stitching = np.zeros(
                (ST_FW_RV.x_list_num, ST_FW_RV.y_list_num, 3))
            for XXXX in range(ST_FW_RV.x_list_num):
                for YYYY in range(ST_FW_RV.y_list_num):
                    shift_FW_stitching[XXXX, YYYY] = FW_RV_stitching_data(
                        XXXX, YYYY)
            ST_FW_RV.shift_FW_stitching = np.array(shift_FW_stitching)
            np.save(os.path.join(ST_FW_RV.stitching_check_path,
                    "FW_RV_stitching_data_np.npy"), ST_FW_RV.FW_RV_stitching_data_np)
            np.save(os.path.join(ST_FW_RV.stitching_check_path,
                    "shift_FW_stitching.npy"), shift_FW_stitching)

            # Determine the FW center (use this stack's coordinates for correction)
            for xxx, yyy in ST_FW_RV.rasen_xy:
                if np.sum(ST_FW_RV.ari_or_nasi(xxx, yyy, "FW"))*np.sum(ST_FW_RV.ari_or_nasi(xxx, yyy, "RV")) > 0:
                    first_fill = xxx, yyy
                    break
            print("FW_center = ", first_fill)

            # Merge at the edge of FW's Z
            merge_z_at_center = ST_FW_RV.FW_z_max
            z_start_at_center = FW_RV_stitching_data(
                first_fill[0], first_fill[1])[2]

            # Function to calculate Z and coordinates for FW (moved to stitching_xy_2060.py)
            ST_FW_RV.z_start_at_center, ST_FW_RV.merge_z_at_center = z_start_at_center, merge_z_at_center
            ST_FW_RV.first_fill, ST_FW_RV.FW_RV_stitching_data = first_fill, FW_RV_stitching_data

            Z_start_at_center = int(FW_RV_stitching_data(
                first_fill[0], first_fill[1])[2])
            FW_z_max = ST_FW_RV.FW_z_max
            
            # Match the MOVIE standard
            make_z_list_FW = list(range(Z_start_at_center, FW_z_max+int(
                Resize_um/ST_FW_RV.scale_z), int(Resize_um/ST_FW_RV.scale_z)))

            # making check image
            ST_FW_RV.white_line_width = 1
            bigs_FW = joblib.Parallel(n_jobs=-1, verbose=1)(joblib.delayed(ST_FW_RV.make_big_image_FW_norm_line)(ZZZ, ResizeUm)
                                                            for ZZZ, ResizeUm in zip(make_z_list_FW, [Resize_um]*len(make_z_list_FW)))
            bigs_FW = np.array(bigs_FW)
            tifffile.imsave(ST_FW_RV.stitching_dst_path+"/FW_RV_st_checking_img_{}um.tif".format(
                str(Resize_um)), (bigs_FW*1000).astype(np.uint16))
            del bigs_FW

        if command["image_making"]["same_size_with_density_image"] == True:
            print("#_making img(same_size_with_density_img)_###_normalized_image_{}um_scale##".format(
                str(Resize_um)))
            # Read the FW_RV stitching that filled the uncalculated areas (calculated only in checker)
            shift_FW_stitching = np.load(os.path.join(
                ST_FW_RV.stitching_check_path, "shift_FW_stitching.npy"))
            ST_FW_RV.FW_RV_stitching_data_np = np.load(os.path.join(
                ST_FW_RV.stitching_check_path, "FW_RV_stitching_data_np.npy"))

            def FW_RV_stitching_data(XX, YY):
                return ST_FW_RV.FW_RV_stitching_data_np[XX, YY]

            # Determine the FW center (use this stack's coordinates for correction)
            for xxx, yyy in ST_FW_RV.rasen_xy:
                if np.sum(ST_FW_RV.ari_or_nasi(xxx, yyy, "FW"))*np.sum(ST_FW_RV.ari_or_nasi(xxx, yyy, "RV")) > 0:
                    first_fill = xxx, yyy
                    break
            print("FW_center = ", first_fill)

            # Merge at the edge of FW's Z
            merge_z_at_center = ST_FW_RV.FW_z_max
            z_start_at_center = FW_RV_stitching_data(
                first_fill[0], first_fill[1])[2]
            
            # Function to calculate Z and coordinates for FW (moved to stitching_xy_2060.py)
            ST_FW_RV.z_start_at_center, ST_FW_RV.merge_z_at_center = z_start_at_center, merge_z_at_center
            ST_FW_RV.first_fill, ST_FW_RV.FW_RV_stitching_data = first_fill, FW_RV_stitching_data

            Z_start_at_center = int(FW_RV_stitching_data(
                first_fill[0], first_fill[1])[2])
            FW_z_max = ST_FW_RV.FW_z_max
            
            print(Z_start_at_center)
            print(FW_z_max)
            print(Resize_um)
            print(ST_FW_RV.scale_z)
            make_z_list_FW = list(range(Z_start_at_center, FW_z_max+int(
                Resize_um/ST_FW_RV.scale_z), int(Resize_um/ST_FW_RV.scale_z)))

            # FW_big_images creation (norm_large)
            bigs_FW = joblib.Parallel(n_jobs=-1, verbose=1)(joblib.delayed(ST_FW_RV.make_big_image_FW_norm)(
                ZZZ, ResizeUm) for ZZZ, ResizeUm in zip(make_z_list_FW, [Resize_um]*len(make_z_list_FW)))
            FWs = np.array(bigs_FW)[1:][::-1]
            del bigs_FW

            # For point_transfer, save FWs_len
            FWs_len = len(FWs)
            print("FWs_len = ", FWs_len)
            np.save(os.path.join(ST_FW_RV.stitching_check_path,
                    "FWs_len.npy"), np.array([FWs_len]))

            # Create RV_big_images (norm_large)
            ST_FW_RV.RV_maisuu_half_5um = np.load(os.path.join(
                ST_FW_RV.stitching_check_path, "RV_maisuu_half_5um.npy"))[0]
            ST_FW_RV.rotation_center_stack_x, ST_FW_RV.rotation_center_stack_y, ST_FW_RV.RV_boundary_hosei = np.load(
                os.path.join(ST_FW_RV.stitching_check_path, "rotation_center_stack_and_RV_boundary_hosei.npy"))
            ST_FW_RV.RV_ants_maisuu065, ST_FW_RV.RV_ants_maisuu065_half = np.load(
                os.path.join(ST_FW_RV.stitching_check_path, "RV_ants_maisuu065.npy"))
            rotation_margin = int(
                ST_FW_RV.RV_maisuu_half_5um * ST_FW_RV.FW_RV_stitching_scale / Resize_um) + 1
            big_z_start = ST_FW_RV.RV_boundary_hosei - \
                rotation_margin * int(Resize_um/ST_FW_RV.scale_z)
            RV_z_max = np.max(np.array([len(glob.glob(os.path.join(stack_path, "*.bin")))
                              for stack_path in glob.glob(os.path.join(ST_xy.src_base_dir_RV, "*/*"))]).astype(np.uint16))
            print(big_z_start)
            print(RV_z_max)
            make_z_list = list(range(big_z_start, 200000+RV_z_max,
                               int(Resize_um/ST_FW_RV.scale_z)))

            bigs_RV = joblib.Parallel(n_jobs=-1, verbose=1)(joblib.delayed(ST_FW_RV.make_big_image_RV_norm)(
                ZZZ, ResizeUm) for ZZZ, ResizeUm in zip(make_z_list, [Resize_um]*len(make_z_list)))

            bigs_RV = np.array(bigs_RV).astype("float32")
            # tifffile.imsave(ST_FW_RV.stitching_check_path+"/before_rotated_20.tif",bigs_RV.astype(np.uint16))
            print(len(bigs_RV))
            print(bigs_RV.shape)
            niiRV_img = tif2nii_no_save(
                bigs_RV, Resize_um, Resize_um, Resize_um)
            del bigs_RV
            antsRV_img = ants.utils.nifti_to_ants(niiRV_img)
            del niiRV_img
            transform_large_img_path = os.path.join(
                ST_FW_RV.stitching_dst_path, "large_image.mat")
            sitk.WriteTransform(ST_FW_RV.Transform_large_Zum(
                big_z_start*ST_FW_RV.scale_z, big_z_start*ST_FW_RV.scale_z), transform_large_img_path)

            W_img = ants.apply_transforms(
                fixed=antsRV_img, moving=antsRV_img, transformlist=transform_large_img_path)
            del antsRV_img
            W_img_np = ST_FW_RV.ants2numpy(W_img)
            del W_img
            RVs = W_img_np[rotation_margin:]
            del W_img_np

            ALLimg = np.concatenate([FWs, RVs], 0)
            del FWs, RVs
            tifffile.imsave(ST_FW_RV.stitching_dst_path+"/norm_stitching_img_{}um.tif".format(
                str(Resize_um)), (ALLimg*1000).astype(np.uint16))
            del ALLimg

    ###############################
    ######### _image_making _######
    ###############################
    if command["image_making"]["normalized_image"] > 0:
        print("#_making img_###_normalized_image_{}um_scale##".format(
            str(command["image_making"]["normalized_image"])))
        # Read shift_FW
        print("load shif_FW and shift_RV...")
        ST_FW_RV.shift_FW = np.load(os.path.join(
            ST_FW_RV.stitching_check_path, "shift_FW.npy"))
        ST_FW_RV.shift_RV = np.load(os.path.join(
            ST_FW_RV.stitching_check_path, "shift_RV.npy"))

        # Read FW_RV stitching that filled uncalculated areas
        shift_FW_stitching = np.load(os.path.join(
            ST_FW_RV.stitching_check_path, "shift_FW_stitching.npy"))
        ST_FW_RV.FW_RV_stitching_data_np = np.load(os.path.join(
            ST_FW_RV.stitching_check_path, "FW_RV_stitching_data_np.npy"))

        # Load FW_RV stitching results
        def FW_RV_stitching_data(XX, YY):
            return ST_FW_RV.FW_RV_stitching_data_np[XX, YY]

        shift_FW_stitching = np.zeros(
            (ST_FW_RV.x_list_num, ST_FW_RV.y_list_num, 3))
        for XXXX in range(ST_FW_RV.x_list_num):
            for YYYY in range(ST_FW_RV.y_list_num):
                shift_FW_stitching[XXXX, YYYY] = FW_RV_stitching_data(
                    XXXX, YYYY)
        ST_FW_RV.shift_FW_stitching = np.array(shift_FW_stitching)

        # Start creating the FW image
        Resize_um = command["image_making"]["normalized_image"]
        # Determine the FW center (use this stack's coordinates for correction)
        for xxx, yyy in ST_FW_RV.rasen_xy:
            if np.sum(ST_FW_RV.ari_or_nasi(xxx, yyy, "FW"))*np.sum(ST_FW_RV.ari_or_nasi(xxx, yyy, "RV")) > 0:
                first_fill = xxx, yyy
                break
        print("FW_center = ", first_fill)

        # Merge at the edge of FW's Z
        merge_z_at_center = ST_FW_RV.FW_z_max
        z_start_at_center = FW_RV_stitching_data(
            first_fill[0], first_fill[1])[2]

        # Function to calculate Z and coordinates for FW (moved to stitching_xy_2060.py)
        ST_FW_RV.z_start_at_center, ST_FW_RV.merge_z_at_center = z_start_at_center, merge_z_at_center
        ST_FW_RV.first_fill, ST_FW_RV.FW_RV_stitching_data = first_fill, FW_RV_stitching_data

        Z_start_at_center = int(FW_RV_stitching_data(
            first_fill[0], first_fill[1])[2])
        FW_z_max = ST_FW_RV.FW_z_max
        make_z_list_FW = list(range(Z_start_at_center, FW_z_max +
                              int(Resize_um/ST_FW_RV.scale_z), int(Resize_um/ST_FW_RV.scale_z)))

        # FW_big_images creation (norm_large)
        bigs_FW = joblib.Parallel(n_jobs=-1, verbose=1)(joblib.delayed(ST_FW_RV.make_big_image_FW_norm)(
            ZZZ, Resize_um) for ZZZ, Resize_um in zip(make_z_list_FW, [Resize_um]*len(make_z_list_FW)))
        FWs = np.array(bigs_FW)[1:][::-1]
        del bigs_FW

        # RV_big_images creation (norm_large)
        ST_FW_RV.RV_maisuu_half_5um = np.load(os.path.join(
            ST_FW_RV.stitching_check_path, "RV_maisuu_half_5um.npy"))[0]
        ST_FW_RV.rotation_center_stack_x, ST_FW_RV.rotation_center_stack_y, ST_FW_RV.RV_boundary_hosei = np.load(
            os.path.join(ST_FW_RV.stitching_check_path, "rotation_center_stack_and_RV_boundary_hosei.npy"))
        ST_FW_RV.RV_ants_maisuu065, ST_FW_RV.RV_ants_maisuu065_half = np.load(
            os.path.join(ST_FW_RV.stitching_check_path, "RV_ants_maisuu065.npy"))
        rotation_margin = int(ST_FW_RV.RV_maisuu_half_5um *
                              ST_FW_RV.FW_RV_stitching_scale / Resize_um) + 1
        big_z_start = ST_FW_RV.RV_boundary_hosei - \
            rotation_margin * int(Resize_um/ST_FW_RV.scale_z)
        RV_z_max = np.max(np.array([len(glob.glob(os.path.join(stack_path, "*.bin")))
                          for stack_path in glob.glob(os.path.join(ST_xy.src_base_dir_RV, "*/*"))]).astype(np.uint16))
        print(big_z_start)
        print(RV_z_max)
        make_z_list = list(range(big_z_start, 200000+RV_z_max,
                           int(Resize_um/ST_FW_RV.scale_z)))

        bigs_RV = joblib.Parallel(n_jobs=-1, verbose=1)(joblib.delayed(ST_FW_RV.make_big_image_RV_norm)(
            ZZZ, Resize_um) for ZZZ, ResizeUm in zip(make_z_list, [Resize_um]*len(make_z_list)))

        bigs_RV = np.array(bigs_RV).astype("float32")
        niiRV_img = tif2nii_no_save(bigs_RV, Resize_um, Resize_um, Resize_um)
        del bigs_RV
        antsRV_img = ants.utils.nifti_to_ants(niiRV_img)
        del niiRV_img
        transform_large_img_path = os.path.join(
            ST_FW_RV.stitching_dst_path, "large_image.mat")
        sitk.WriteTransform(ST_FW_RV.Transform_large_Zum(
            big_z_start*ST_FW_RV.scale_z, big_z_start*ST_FW_RV.scale_z), transform_large_img_path)

        W_img = ants.apply_transforms(
            fixed=antsRV_img, moving=antsRV_img, transformlist=transform_large_img_path)
        del antsRV_img
        W_img_np = ST_FW_RV.ants2numpy(W_img)
        del W_img
        RVs = W_img_np[rotation_margin:]
        del W_img_np

        ALLimg = np.concatenate([FWs, RVs], 0)
        del FWs, RVs
        tifffile.imsave(ST_FW_RV.stitching_dst_path+"/norm_stitching_img_{}um.tif".format(
            str(Resize_um)), (ALLimg*1000).astype(np.uint16))
        del ALLimg
        print("making img_###_normalized_image_{}um_scale_finished!!".format(
            str(command["image_making"]["normalized_image"])))

    if command["image_making"]["average_image"] > 0:
        print("#_making img_###_average_image_{}um_scale##".format(
            str(command["image_making"]["average_image"])))
        # Read shift_FW
        print("load shif_FW and shift_RV...")
        ST_FW_RV.shift_FW = np.load(os.path.join(
            ST_FW_RV.stitching_check_path, "shift_FW.npy"))
        ST_FW_RV.shift_RV = np.load(os.path.join(
            ST_FW_RV.stitching_check_path, "shift_RV.npy"))

        # Read FW_RV stitching that filled uncalculated areas
        shift_FW_stitching = np.load(os.path.join(
            ST_FW_RV.stitching_check_path, "shift_FW_stitching.npy"))
        ST_FW_RV.FW_RV_stitching_data_np = np.load(os.path.join(
            ST_FW_RV.stitching_check_path, "FW_RV_stitching_data_np.npy"))
        # Load FW_RV stitching results

        def FW_RV_stitching_data(XX, YY):
            return ST_FW_RV.FW_RV_stitching_data_np[XX, YY]

        shift_FW_stitching = np.zeros(
            (ST_FW_RV.x_list_num, ST_FW_RV.y_list_num, 3))
        for XXXX in range(ST_FW_RV.x_list_num):
            for YYYY in range(ST_FW_RV.y_list_num):
                shift_FW_stitching[XXXX, YYYY] = FW_RV_stitching_data(
                    XXXX, YYYY)
        ST_FW_RV.shift_FW_stitching = np.array(shift_FW_stitching)

        # Start creating the FW image
        Resize_um = command["image_making"]["average_image"]
        # Determine the FW center (use this stack's coordinates for correction)
        for xxx, yyy in ST_FW_RV.rasen_xy:
            if np.sum(ST_FW_RV.ari_or_nasi(xxx, yyy, "FW"))*np.sum(ST_FW_RV.ari_or_nasi(xxx, yyy, "RV")) > 0:
                first_fill = xxx, yyy
                break
        print("FW_center = ", first_fill)

        # Merge at the edge of FW's Z
        merge_z_at_center = ST_FW_RV.FW_z_max
        z_start_at_center = FW_RV_stitching_data(
            first_fill[0], first_fill[1])[2]

        # Function to calculate Z and coordinates for FW (moved to stitching_xy_2060.py)
        ST_FW_RV.z_start_at_center, ST_FW_RV.merge_z_at_center = z_start_at_center, merge_z_at_center
        ST_FW_RV.first_fill, ST_FW_RV.FW_RV_stitching_data = first_fill, FW_RV_stitching_data

        Z_start_at_center = int(FW_RV_stitching_data(
            first_fill[0], first_fill[1])[2])
        FW_z_max = ST_FW_RV.FW_z_max
        make_z_list_FW = list(range(Z_start_at_center, FW_z_max +
                              int(Resize_um/ST_FW_RV.scale_z), int(Resize_um/ST_FW_RV.scale_z)))

        # FW_big_images creation (norm_large)
        bigs_FW = joblib.Parallel(n_jobs=-1, verbose=1)(joblib.delayed(ST_FW_RV.make_big_image_FW)(
            ZZZ, Resize_um) for ZZZ, ResizeUm in zip(make_z_list_FW, [Resize_um]*len(make_z_list_FW)))
        FWs = np.array(bigs_FW)[1:][::-1]
        del bigs_FW

        # RV_big_images creation (norm_large)
        ST_FW_RV.RV_maisuu_half_5um = np.load(os.path.join(
            ST_FW_RV.stitching_check_path, "RV_maisuu_half_5um.npy"))[0]
        ST_FW_RV.rotation_center_stack_x, ST_FW_RV.rotation_center_stack_y, ST_FW_RV.RV_boundary_hosei = np.load(
            os.path.join(ST_FW_RV.stitching_check_path, "rotation_center_stack_and_RV_boundary_hosei.npy"))
        ST_FW_RV.RV_ants_maisuu065, ST_FW_RV.RV_ants_maisuu065_half = np.load(
            os.path.join(ST_FW_RV.stitching_check_path, "RV_ants_maisuu065.npy"))
        rotation_margin = int(ST_FW_RV.RV_maisuu_half_5um *
                              ST_FW_RV.FW_RV_stitching_scale / Resize_um) + 1
        big_z_start = ST_FW_RV.RV_boundary_hosei - \
            rotation_margin * int(Resize_um/ST_FW_RV.scale_z)
        RV_z_max = np.max(np.array([len(glob.glob(os.path.join(stack_path, "*.bin")))
                          for stack_path in glob.glob(os.path.join(ST_xy.src_base_dir_RV, "*/*"))]).astype(np.uint16))
        print(big_z_start)
        print(RV_z_max)
        make_z_list = list(range(big_z_start, 200000+RV_z_max,
                           int(Resize_um/ST_FW_RV.scale_z)))

        bigs_RV = joblib.Parallel(n_jobs=-1, verbose=1)(joblib.delayed(ST_FW_RV.make_big_image_RV)(
            ZZZ, Resize_um) for ZZZ, ResizeUm in zip(make_z_list, [Resize_um]*len(make_z_list)))

        bigs_RV = np.array(bigs_RV).astype("float32")
        niiRV_img = tif2nii_no_save(bigs_RV, Resize_um, Resize_um, Resize_um)
        del bigs_RV
        antsRV_img = ants.utils.nifti_to_ants(niiRV_img)
        del niiRV_img
        transform_large_img_path = os.path.join(
            ST_FW_RV.stitching_dst_path, "large_image.mat")
        sitk.WriteTransform(ST_FW_RV.Transform_large_Zum(
            big_z_start*ST_FW_RV.scale_z, big_z_start*ST_FW_RV.scale_z), transform_large_img_path)

        W_img = ants.apply_transforms(
            fixed=antsRV_img, moving=antsRV_img, transformlist=transform_large_img_path)
        del antsRV_img
        W_img_np = ST_FW_RV.ants2numpy(W_img)
        del W_img
        RVs = W_img_np[rotation_margin:]
        del W_img_np

        ALLimg = np.concatenate([FWs, RVs], 0)
        del FWs, RVs
        tifffile.imsave(ST_FW_RV.stitching_dst_path+"/average_stitching_img_{}um.tif".format(
            str(Resize_um)), (ALLimg).astype(np.uint16))
        del ALLimg
        print("making img_###_average_image_{}um_scale_finished!!".format(
            str(command["image_making"]["average_image"])))

    ###############################
    ####### _point_transfer _######
    ###############################
    if command["point_transfer"]:
        print("#######_point_transfer _#######")
        # Read rotation_center and RV_hosei
        ST_FW_RV.rotation_center_stack_x, ST_FW_RV.rotation_center_stack_y, ST_FW_RV.RV_boundary_hosei = np.load(
            os.path.join(ST_FW_RV.stitching_check_path, "rotation_center_stack_and_RV_boundary_hosei.npy"))
        ST_FW_RV.RV_ants_maisuu065, ST_FW_RV.RV_ants_maisuu065_half = np.load(
            os.path.join(ST_FW_RV.stitching_check_path, "RV_ants_maisuu065.npy"))
        # From here, move FW points
        # The boundary z-position (um) from the perspective of FW
        Resize_um = command["density_image_voxel_size_um"]
        FWs_len = np.load(os.path.join(
            ST_FW_RV.stitching_check_path, "FWs_len.npy"))[0]
        ST_FW_RV.boundary_um = FWs_len*Resize_um

        # Determine the FW center (use this stack's coordinates for correction)
        for xxx, yyy in ST_FW_RV.rasen_xy:
            if np.sum(ST_FW_RV.ari_or_nasi(xxx, yyy, "FW"))*np.sum(ST_FW_RV.ari_or_nasi(xxx, yyy, "RV")) > 0:
                first_fill = xxx, yyy
                break
        print("FW_center = ", first_fill)

        # Read shift_FW
        print("load shif_FW and shift_RV...")
        ST_FW_RV.shift_FW = np.load(os.path.join(
            ST_FW_RV.stitching_check_path, "shift_FW.npy"))
        ST_FW_RV.shift_RV = np.load(os.path.join(
            ST_FW_RV.stitching_check_path, "shift_RV.npy"))

        print("load FW points...")
        # Read FW_RV stitching that filled uncalculated areas
        shift_FW_stitching = np.load(os.path.join(
            ST_FW_RV.stitching_check_path, "shift_FW_stitching.npy"))
        ST_FW_RV.FW_RV_stitching_data_np = np.load(os.path.join(
            ST_FW_RV.stitching_check_path, "FW_RV_stitching_data_np.npy"))
        FW_RV_stitching_data = np.load(os.path.join(
            ST_FW_RV.stitching_check_path, "FW_RV_stitching_data_np.npy"))

        # Load FW_RV stitching results
        def FW_RV_stitching_data(XX, YY):
            return ST_FW_RV.FW_RV_stitching_data_np[XX, YY]

        shift_FW_stitching = np.zeros(
            (ST_FW_RV.x_list_num, ST_FW_RV.y_list_num, 3))
        for XXXX in range(ST_FW_RV.x_list_num):
            for YYYY in range(ST_FW_RV.y_list_num):
                shift_FW_stitching[XXXX, YYYY] = FW_RV_stitching_data(
                    XXXX, YYYY)
        ST_FW_RV.shift_FW_stitching = np.array(shift_FW_stitching)

        # z is relative to the z of first_fill's stack; x, y convert to stitching coordinates
        ST_FW_RV.shift_FW_hosei = ST_FW_RV.shift_FW - ST_FW_RV.shift_FW[first_fill[0], first_fill[1]] + np.array(
            [FW_RV_stitching_data(first_fill[0], first_fill[1])[0], FW_RV_stitching_data(first_fill[0], first_fill[1])[1], 0])

        # Merge at the edge of FW's Z
        merge_z_at_center = ST_FW_RV.FW_z_max

        def FW_LeftUp_marge(xxx, yyy):  # The upper-left global coordinate of the boundary, above the merge
            if xxx == 0:
                if yyy == 0:
                    return np.array([ST_FW_RV.x_overlap, ST_FW_RV.y_overlap])
                if yyy != 0 and yyy != ST_FW_RV.y_list_num:
                    x_coordinate = max(
                        ST_FW_RV.shift_FW_hosei[xxx, yyy][0], ST_FW_RV.shift_FW_hosei[xxx, yyy-1][0]) + ST_FW_RV.x_overlap
                    y_coordinate = (
                        (ST_FW_RV.shift_FW_hosei[xxx, yyy-1][1] + ST_FW_RV.y_pixel) + ST_FW_RV.shift_FW_hosei[xxx, yyy][1])/2
                    return np.array([x_coordinate, y_coordinate])
            if yyy == 0 and xxx != ST_FW_RV.x_list_num:
                x_coordinate = (
                    (ST_FW_RV.shift_FW_hosei[xxx-1, yyy][0] + ST_FW_RV.x_pixel) + ST_FW_RV.shift_FW_hosei[xxx, yyy][0])/2
                y_coordinate = max(
                    ST_FW_RV.shift_FW_hosei[xxx, yyy][1], ST_FW_RV.shift_FW_hosei[xxx-1, yyy][1]) + ST_FW_RV.y_overlap
                return np.array([x_coordinate, ST_FW_RV.x_overlap])

            if xxx == ST_FW_RV.x_list_num:
                if yyy == ST_FW_RV.y_list_num:
                    return ST_FW_RV.shift_FW_hosei[ST_FW_RV.x_list_num-1, ST_FW_RV.y_list_num-1][:2] + np.array([ST_FW_RV.x_pixel, ST_FW_RV.y_pixel]) - np.array([ST_FW_RV.x_overlap, ST_FW_RV.y_overlap])
                if yyy == 0:
                    x_coordinate = ST_FW_RV.shift_FW_hosei[xxx-1,
                                                           yyy][0] + ST_FW_RV.x_pixel - ST_FW_RV.x_overlap
                    y_coordinate = ST_FW_RV.shift_FW_hosei[xxx -
                                                           1, yyy][1] + ST_FW_RV.y_overlap
                    return np.array([x_coordinate, y_coordinate])
                if yyy != ST_FW_RV.y_list_num and yyy != 0:
                    x_coordinate = min(
                        ST_FW_RV.shift_FW_hosei[xxx-1, yyy][0], ST_FW_RV.shift_FW_hosei[xxx-1, yyy-1][0]) + ST_FW_RV.x_pixel-ST_FW_RV.x_overlap
                    y_coordinate = (
                        (ST_FW_RV.shift_FW_hosei[xxx-1, yyy-1][1] + ST_FW_RV.y_pixel) + ST_FW_RV.shift_FW_hosei[xxx-1, yyy][1])/2
                    return np.array([x_coordinate, y_coordinate])
            if yyy == ST_FW_RV.y_list_num:
                if xxx == 0:
                    x_coordinate = ST_FW_RV.shift_FW_hosei[xxx,
                                                           yyy-1][0] + ST_FW_RV.x_overlap
                    y_coordinate = ST_FW_RV.shift_FW_hosei[xxx, yyy -
                                                           1][1] + ST_FW_RV.y_pixel - ST_FW_RV.y_overlap
                    return np.array([x_coordinate, y_coordinate])
                if xxx != 0:
                    x_coordinate = (
                        (ST_FW_RV.shift_FW_hosei[xxx-1, yyy-1][0] + ST_FW_RV.x_pixel) + ST_FW_RV.shift_FW_hosei[xxx, yyy-1][0])/2
                    y_coordinate = min(
                        ST_FW_RV.shift_FW_hosei[xxx-1, yyy-1][1], ST_FW_RV.shift_FW_hosei[xxx, yyy-1][1]) + ST_FW_RV.y_pixel - ST_FW_RV.y_overlap
                    return np.array([x_coordinate, y_coordinate])

            x_coordinate = (max(ST_FW_RV.shift_FW_hosei[xxx, yyy][0], ST_FW_RV.shift_FW_hosei[xxx, yyy-1][0]) + min(
                ST_FW_RV.shift_FW_hosei[xxx-1, yyy][0]+ST_FW_RV.x_pixel, ST_FW_RV.shift_FW_hosei[xxx-1, yyy-1][0]+ST_FW_RV.x_pixel))/2
            y_coordinate = (max(ST_FW_RV.shift_FW_hosei[xxx, yyy][1], ST_FW_RV.shift_FW_hosei[xxx-1, yyy][1]) + min(
                ST_FW_RV.shift_FW_hosei[xxx, yyy-1][1]+ST_FW_RV.x_pixel, ST_FW_RV.shift_FW_hosei[xxx-1, yyy-1][1]+ST_FW_RV.y_pixel))/2
            return np.array([x_coordinate, y_coordinate])

        def FW_LeftUp_stitching(xxx, yyy):  # The upper-left global coordinate of the boundary, stitching interface
            if xxx == 0:
                if yyy == 0:
                    return np.array([ST_FW_RV.x_overlap, ST_FW_RV.y_overlap])
                if yyy != 0 and yyy != ST_FW_RV.y_list_num:
                    x_coordinate = max(
                        ST_FW_RV.shift_FW_stitching[xxx, yyy][0], ST_FW_RV.shift_FW_stitching[xxx, yyy-1][0]) + ST_FW_RV.x_overlap
                    y_coordinate = (
                        (ST_FW_RV.shift_FW_stitching[xxx, yyy-1][1] + ST_FW_RV.y_pixel) + ST_FW_RV.shift_FW_stitching[xxx, yyy][1])/2
                    return np.array([x_coordinate, y_coordinate])
            if yyy == 0 and xxx != ST_FW_RV.x_list_num:
                x_coordinate = (
                    (ST_FW_RV.shift_FW_stitching[xxx-1, yyy][0] + ST_FW_RV.x_pixel) + ST_FW_RV.shift_FW_stitching[xxx, yyy][0])/2
                y_coordinate = max(
                    ST_FW_RV.shift_FW_stitching[xxx, yyy][1], ST_FW_RV.shift_FW_stitching[xxx-1, yyy][1]) + ST_FW_RV.y_overlap
                return np.array([x_coordinate, ST_FW_RV.x_overlap])

            if xxx == ST_FW_RV.x_list_num:
                if yyy == ST_FW_RV.y_list_num:
                    return ST_FW_RV.shift_FW_stitching[ST_FW_RV.x_list_num-1, ST_FW_RV.y_list_num-1][:2] + np.array([ST_FW_RV.x_pixel, ST_FW_RV.y_pixel]) - np.array([ST_FW_RV.x_overlap, ST_FW_RV.y_overlap])
                if yyy == 0:
                    x_coordinate = ST_FW_RV.shift_FW_stitching[xxx-1,
                                                               yyy][0] + ST_FW_RV.x_pixel - ST_FW_RV.x_overlap
                    y_coordinate = ST_FW_RV.shift_FW_stitching[xxx -
                                                               1, yyy][1] + ST_FW_RV.y_overlap
                    return np.array([x_coordinate, y_coordinate])
                if yyy != ST_FW_RV.y_list_num and yyy != 0:
                    x_coordinate = min(
                        ST_FW_RV.shift_FW_stitching[xxx-1, yyy][0], ST_FW_RV.shift_FW_stitching[xxx-1, yyy-1][0]) + ST_FW_RV.x_pixel-ST_FW_RV.x_overlap
                    y_coordinate = (
                        (ST_FW_RV.shift_FW_stitching[xxx-1, yyy-1][1] + ST_FW_RV.y_pixel) + ST_FW_RV.shift_FW_stitching[xxx-1, yyy][1])/2
                    return np.array([x_coordinate, y_coordinate])
            if yyy == ST_FW_RV.y_list_num:
                if xxx == 0:
                    x_coordinate = ST_FW_RV.shift_FW_stitching[xxx,
                                                               yyy-1][0] + ST_FW_RV.x_overlap
                    y_coordinate = ST_FW_RV.shift_FW_stitching[xxx,
                                                               yyy-1][1] + ST_FW_RV.y_pixel - ST_FW_RV.y_overlap
                    return np.array([x_coordinate, y_coordinate])
                if xxx != 0:
                    x_coordinate = (
                        (ST_FW_RV.shift_FW_stitching[xxx-1, yyy-1][0] + ST_FW_RV.x_pixel) + ST_FW_RV.shift_FW_stitching[xxx, yyy-1][0])/2
                    y_coordinate = min(
                        ST_FW_RV.shift_FW_stitching[xxx-1, yyy-1][1], ST_FW_RV.shift_FW_stitching[xxx, yyy-1][1]) + ST_FW_RV.y_pixel - ST_FW_RV.y_overlap
                    return np.array([x_coordinate, y_coordinate])

            x_coordinate = (max(ST_FW_RV.shift_FW_stitching[xxx, yyy][0], ST_FW_RV.shift_FW_stitching[xxx, yyy-1][0]) + min(
                ST_FW_RV.shift_FW_stitching[xxx-1, yyy][0]+ST_FW_RV.x_pixel, ST_FW_RV.shift_FW_stitching[xxx-1, yyy-1][0]+ST_FW_RV.x_pixel))/2
            y_coordinate = (max(ST_FW_RV.shift_FW_stitching[xxx, yyy][1], ST_FW_RV.shift_FW_stitching[xxx-1, yyy][1]) + min(
                ST_FW_RV.shift_FW_stitching[xxx, yyy-1][1]+ST_FW_RV.x_pixel, ST_FW_RV.shift_FW_stitching[xxx-1, yyy-1][1]+ST_FW_RV.y_pixel))/2
            return np.array([x_coordinate, y_coordinate])

        def FW_LeftUp(xxx, yyy, zzz):  # The upper-left global coordinate of the boundary, zzz: global coordinate
            point_ratio = zzz / merge_z_at_center
            return FW_LeftUp_stitching(xxx, yyy) + (FW_LeftUp_marge(xxx, yyy) - FW_LeftUp_stitching(xxx, yyy)) * point_ratio

        def global_FW(XX, YY):
            if ST_FW_RV.load_point_data(XX, YY, "FW").size == 0:
                return np.empty((3, 0))
            else:
                FW_points_local = ST_FW_RV.load_point_data(XX, YY, "FW")
                print(XX, YY, len(ST_FW_RV.load_point_data(XX, YY, "FW")))
                point_ratio = (FW_points_local[2] - FW_RV_stitching_data(XX, YY)[2])/(
                    merge_z_at_center + ST_FW_RV.shift_FW_hosei[XX, YY, 2] - FW_RV_stitching_data(XX, YY)[2])

                stack_hidariue_marge = np.array(
                    [ST_FW_RV.shift_FW_hosei[XX, YY, 0], ST_FW_RV.shift_FW_hosei[XX, YY, 1], merge_z_at_center])
                stack_hidariue_boundary = np.array([FW_RV_stitching_data(
                    XX, YY)[0], FW_RV_stitching_data(XX, YY)[1], FW_RV_stitching_data(3, 3)[2]])

                FW_points_global_x = point_ratio*(stack_hidariue_marge - stack_hidariue_boundary)[
                    0] + stack_hidariue_boundary[0] + FW_points_local[0]
                FW_points_global_y = point_ratio*(stack_hidariue_marge - stack_hidariue_boundary)[
                    1] + stack_hidariue_boundary[1] + FW_points_local[1]
                FW_points_global_z = point_ratio * \
                    (stack_hidariue_marge - stack_hidariue_boundary)[2]
                return np.array([FW_points_global_x, FW_points_global_y, FW_points_global_z])
        
        def global_FW_all(XX, YY):
            if ST_FW_RV.load_point_data_all(XX, YY, "FW").size == 0:
                #print("emp")
                return np.empty((3, 0))
            else:
                FW_points_local = ST_FW_RV.load_point_data_all(XX, YY, "FW")
                print(XX, YY, len(ST_FW_RV.load_point_data_all(XX, YY, "FW")))
                point_ratio = (FW_points_local[2] - FW_RV_stitching_data(XX, YY)[2])/(
                    merge_z_at_center + ST_FW_RV.shift_FW_hosei[XX, YY, 2] - FW_RV_stitching_data(XX, YY)[2])

                stack_hidariue_marge = np.array(
                    [ST_FW_RV.shift_FW_hosei[XX, YY, 0], ST_FW_RV.shift_FW_hosei[XX, YY, 1], merge_z_at_center])
                stack_hidariue_boundary = np.array([FW_RV_stitching_data(
                    XX, YY)[0], FW_RV_stitching_data(XX, YY)[1], FW_RV_stitching_data(3, 3)[2]])

                FW_points_global_x = point_ratio*(stack_hidariue_marge - stack_hidariue_boundary)[
                    0] + stack_hidariue_boundary[0] + FW_points_local[0]
                FW_points_global_y = point_ratio*(stack_hidariue_marge - stack_hidariue_boundary)[
                    1] + stack_hidariue_boundary[1] + FW_points_local[1]
                FW_points_global_z = point_ratio * (stack_hidariue_marge - stack_hidariue_boundary)[2]
                return np.array([FW_points_global_x, FW_points_global_y, FW_points_global_z])

        # x,y :pixel, z: miasuu
        def global_FW_clipped(xxx, yyy):           
            if ST_FW_RV.load_point_data(xxx, yyy, "FW").size == 0:
                return np.empty((3, 0))
            else:
                FW_points_stack = global_FW(xxx, yyy)
                FW_LeftUp_stack_X0_Y0 = np.array(
                    [FW_LeftUp(xxx, yyy, zzzzz) for zzzzz in FW_points_stack[2]])
                FW_LeftUp_stack_X1_Y0 = np.array(
                    [FW_LeftUp(xxx+1, yyy, zzzzz) for zzzzz in FW_points_stack[2]])
                FW_LeftUp_stack_X0_Y1 = np.array(
                    [FW_LeftUp(xxx, yyy+1, zzzzz) for zzzzz in FW_points_stack[2]])
                FW_LeftUp_stack_X1_Y1 = np.array(
                    [FW_LeftUp(xxx+1, yyy+1, zzzzz) for zzzzz in FW_points_stack[2]])

                # up, line_formula_y
                x1, x2, y1, y2 = FW_LeftUp_stack_X0_Y0.T[0], FW_LeftUp_stack_X1_Y0.T[0], FW_LeftUp_stack_X0_Y0.T[1], FW_LeftUp_stack_X1_Y0.T[1]
                a_up = (y2-y1)/(x2-x1)
                b_up = (-x1*y2 + x2*y1)/(x2-x1)
                point_clf0 = (FW_points_stack[1]
                              > a_up*FW_points_stack[0] + b_up)

                # down, line_formula_y
                x1, x2, y1, y2 = FW_LeftUp_stack_X0_Y1.T[0], FW_LeftUp_stack_X1_Y1.T[0], FW_LeftUp_stack_X0_Y1.T[1], FW_LeftUp_stack_X1_Y1.T[1]
                a_down = (y2-y1)/(x2-x1)
                b_down = (-x1*y2 + x2*y1)/(x2-x1)
                point_clf1 = (
                    FW_points_stack[1] < a_down * FW_points_stack[0] + b_down)

                # left, line_formula_x
                x1, x2, y1, y2 = FW_LeftUp_stack_X0_Y0.T[0], FW_LeftUp_stack_X0_Y1.T[0], FW_LeftUp_stack_X0_Y0.T[1], FW_LeftUp_stack_X0_Y1.T[1]
                a_left = (x2-x1)/(y2-y1)
                b_left = (-y1*x2 + y2*x1)/(y2-y1)
                point_clf2 = (
                    FW_points_stack[0] > a_left * FW_points_stack[1] + b_left)

                # right, line_formula_x
                x1, x2, y1, y2 = FW_LeftUp_stack_X1_Y0.T[0], FW_LeftUp_stack_X1_Y1.T[0], FW_LeftUp_stack_X1_Y0.T[1], FW_LeftUp_stack_X1_Y1.T[1]
                a_right = (x2-x1)/(y2-y1)
                b_right = (-y1*x2 + y2*x1)/(y2-y1)
                point_clf3 = (
                    FW_points_stack[0] < a_right * FW_points_stack[1] + b_right)

                # #0 or below is excluded
                point_clf4 = FW_points_stack[2] > 0

                point_clf = point_clf0*point_clf1*point_clf2*point_clf3*point_clf4
                return FW_points_stack.T[point_clf].T
        
        # x,y :pixel, z: miasuu
        def global_FW_clipped_all (xxx, yyy):
            FW_points_stack = np.empty((3, 0))
            FW_points_um = np.empty((3, 0))
            if ST_FW_RV.load_point_data_all(xxx, yyy, "FW").size == 0:
                print("emp_c")
                #return np.empty((3, 0))
            else:
                FW_points_stack = global_FW_all(xxx, yyy)       
                FW_LeftUp_stack_X0_Y0 = np.array(
                    [FW_LeftUp(xxx, yyy, zzzzz) for zzzzz in FW_points_stack[2]])
                FW_LeftUp_stack_X1_Y0 = np.array(
                    [FW_LeftUp(xxx+1, yyy, zzzzz) for zzzzz in FW_points_stack[2]])
                FW_LeftUp_stack_X0_Y1 = np.array(
                    [FW_LeftUp(xxx, yyy+1, zzzzz) for zzzzz in FW_points_stack[2]])
                FW_LeftUp_stack_X1_Y1 = np.array(
                    [FW_LeftUp(xxx+1, yyy+1, zzzzz) for zzzzz in FW_points_stack[2]])

                # up, line_formula_y
                x1, x2, y1, y2 = FW_LeftUp_stack_X0_Y0.T[0], FW_LeftUp_stack_X1_Y0.T[0], FW_LeftUp_stack_X0_Y0.T[1], FW_LeftUp_stack_X1_Y0.T[1]
                a_up = (y2-y1)/(x2-x1)
                b_up = (-x1*y2 + x2*y1)/(x2-x1)
                point_clf0 = (FW_points_stack[1]
                              > a_up*FW_points_stack[0] + b_up)

                # down, line_formula_y
                x1, x2, y1, y2 = FW_LeftUp_stack_X0_Y1.T[0], FW_LeftUp_stack_X1_Y1.T[0], FW_LeftUp_stack_X0_Y1.T[1], FW_LeftUp_stack_X1_Y1.T[1]
                a_down = (y2-y1)/(x2-x1)
                b_down = (-x1*y2 + x2*y1)/(x2-x1)
                point_clf1 = (
                    FW_points_stack[1] < a_down * FW_points_stack[0] + b_down)

                # left, line_formula_x
                x1, x2, y1, y2 = FW_LeftUp_stack_X0_Y0.T[0], FW_LeftUp_stack_X0_Y1.T[0], FW_LeftUp_stack_X0_Y0.T[1], FW_LeftUp_stack_X0_Y1.T[1]
                a_left = (x2-x1)/(y2-y1)
                b_left = (-y1*x2 + y2*x1)/(y2-y1)
                point_clf2 = (
                    FW_points_stack[0] > a_left * FW_points_stack[1] + b_left)

                # right, line_formula_x
                x1, x2, y1, y2 = FW_LeftUp_stack_X1_Y0.T[0], FW_LeftUp_stack_X1_Y1.T[0], FW_LeftUp_stack_X1_Y0.T[1], FW_LeftUp_stack_X1_Y1.T[1]
                a_right = (x2-x1)/(y2-y1)
                b_right = (-y1*x2 + y2*x1)/(y2-y1)
                point_clf3 = (
                    FW_points_stack[0] < a_right * FW_points_stack[1] + b_right)

                # #0 or below is excluded
                point_clf4 = FW_points_stack[2] > 0

                point_clf = point_clf0*point_clf1*point_clf2*point_clf3*point_clf4
                # point_clf is applied. negative points are set to [0,0,0]
                FW_points_stack.T[point_clf < 1] = [0,0, ST_FW_RV.boundary_um / ST_FW_RV.scale_z]
                FW_points_um = np.array([FW_points_stack[0]*ST_FW_RV.scale_x, FW_points_stack[1]
                                        * ST_FW_RV.scale_y,  ST_FW_RV.boundary_um - FW_points_stack[2] * ST_FW_RV.scale_z])
            ST_FW_RV.save_stitched_stack_data(xxx, yyy, "FW", FW_points_um)
            return FW_points_um # Meaningless return

        if 1==0:
            ut1 = time.time()
            all_data_tmp = joblib.Parallel(n_jobs=-1,  verbose=1)([joblib.delayed(global_FW_clipped)(
                xxxxxx, yyyyyy) for xxxxxx in range(ST_FW_RV.x_list_num) for yyyyyy in range(ST_FW_RV.y_list_num)])
            print("transfer FW points takes ", time.time()-ut1, " s")

            FW_all_points = np.concatenate(all_data_tmp, 1)
            FW_all_points_um = np.array([FW_all_points[0]*ST_FW_RV.scale_x, FW_all_points[1]
                                        * ST_FW_RV.scale_y,  ST_FW_RV.boundary_um - FW_all_points[2] * ST_FW_RV.scale_z])

            print("load RV points...")
            # Load and rotate RV points
            ST_FW_RV.rotation_of_all_RV_points()

            # Move RV points
            RV_all_points_um_tmp = ST_FW_RV.roteted_RV_points.T[ST_FW_RV.roteted_RV_points[2] > 0].T
            RV_all_points_um = np.array(
                [RV_all_points_um_tmp[0], RV_all_points_um_tmp[1], ST_FW_RV.boundary_um+RV_all_points_um_tmp[2]])

            # Combine FW_RV points
            all_points_um = np.concatenate([FW_all_points_um, RV_all_points_um], 1)
            np.save(os.path.join(ST_FW_RV.stitching_dst_path,
                    "all_points_um.npy"), all_points_um)
            print("total cell: ", len(all_points_um[0]))
            np.save(os.path.join(ST_FW_RV.stitching_dst_path, "total_cell_num_{}.npy".format(
                str(len(all_points_um[0])).zfill(12))), np.zeros(1))

            # making_density_img         
            
            list_x, list_y, list_z = all_points_um/Resize_um
            ALLimg = tifffile.imread(
                ST_FW_RV.stitching_dst_path+"/norm_stitching_img_{}um.tif".format(str(Resize_um)))
            size_z, size_y, size_x = ALLimg.shape
            del ALLimg
        
        
            print("making density image...")


            d = make_density_image(list_x, list_y, list_z, size_x, size_y, size_z)
            tifffile.imsave(ST_FW_RV.stitching_dst_path +
                            "/density_img_{}um.tif".format(str(Resize_um)), (d).astype(np.uint16))
            print("density_image_saved!")

            
        
        print("save stitched xyz coordinate start")
        ut1 = time.time()

        with Parallel(n_jobs=40,  verbose=1) as parallel:
            all_data_tmp2 = parallel([delayed(global_FW_clipped_all)(xxxxxx, yyyyyy) 
                                    for xxxxxx in range(ST_FW_RV.x_list_num) 
                                    for yyyyyy in range(ST_FW_RV.y_list_num)])

        print("transfer FW2 points takes ", time.time()-ut1, " s")
        
        ut1 = time.time()

        with Parallel(n_jobs=40,  verbose=1) as parallel:
            all_data_tmp3 = parallel([delayed(ST_FW_RV.save_stitched_RV_stack_data_all)(
                xxxxxx, yyyyyy, ST_FW_RV.boundary_um) for xxxxxx in range(ST_FW_RV.x_list_num) for yyyyyy in range(ST_FW_RV.y_list_num)])


        print("transfer RV2 points takes ", time.time()-ut1, " s")
        
        print("save stitched xyz coordinate end")
```
